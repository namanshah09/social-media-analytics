```python
import codecs
import re
with codecs.open('/content/YouTubeTrailerComments.txt', encoding='ISO-8859-1') as f:
    movie_id=0
    for line in f:
        
        if("MovieID::::" in line):
            movie_id=line.replace("MovieID::::","")
            movie_id= movie_id.replace("\r\n","")
            
        if re.match(r'(\d+)\s+(\d+)\s+(\d+)\s+(\d+)\s+(\d+)',line):
            lines_data=line.split();
            lines_data.insert(0,movie_id)
            print(','.join(lines_data))
```
